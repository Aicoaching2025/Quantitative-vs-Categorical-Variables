```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/titanic.csv'
```

```python
titanic = pd.read_csv(file_path)
```

```python
print(titanic.head())
```

```python
# Compute mean of Fare for survivors and non-survivors
Survived_mean = titanic[titanic['Survived'] == 1]['Fare'].mean()
Died_mean = titanic[titanic['Survived'] == 0]['Fare'].mean()

# Calculate the difference in means
diff_mean = Survived_mean - Died_mean
print(f"Difference in Mean Fare (Survived - Died): {diff_mean:.2f}")

# Compute median of Fare for survivors and non-survivors
Survived_med = np.median(titanic[titanic['Survived'] == 1]['Fare'])
Died_med = np.median(titanic[titanic['Survived'] == 0]['Fare'])

# Calculate the difference in medians
diff_med = Survived_med - Died_med
print(f"Difference in Median Fare (Survived - Died): {diff_med:.2f}")

# Create a Boxplot of Fare by Survival Status
plt.figure(figsize=(10, 12))
sns.boxplot(data=titanic, hue='Survived', y='Fare', hue="coolwarm")
plt.title("Fare Distribution by Survival Status", fontsize=14)
plt.xlabel("Survived (0 = No, 1 = Yes)", fontsize=12)
plt.ylabel("Fare", fontsize=12)
plt.show()
```
